```python
import re
import time
from googletrans import Translator
import string
import requests
from urllib.parse import quote
import pandas as pd
```

```python
data = pd.read_csv("../Sentiment_Analysis_Dataset.csv")
data
```

```python
source_language = "fa"
target_language = "ur"

index = 4400
total_len = len(data)
translations = []
for sentence in data["sentence"][index:]:
    print(f"Translating Text: {index+1}/{total_len+1}")

    try:    
        url_encoded_text = quote(sentence)  
        url = f"https://translate.googleapis.com/translate_a/single?client=gtx&sl={source_language}&tl={target_language}&dt=t&q={url_encoded_text}"
        response = requests.get(url)
        translated = response.json()[0][0][0]
    
        translations.append(translated)   
    except Exception as e:
        print("Translation error:", str(e))
    
    # Add a delay of a few seconds between requests
    time.sleep(1.88)
    index+=1

    if index%100 == 0:
        with open(f"./sentiment/translations_{index}.txt", "w", encoding="utf-8") as f:
            f.write("\n".join(translations))
            translations = []
    
```

```python
with open(f"./sentiment/translations_{index}.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(translations))
    translations = []
```
